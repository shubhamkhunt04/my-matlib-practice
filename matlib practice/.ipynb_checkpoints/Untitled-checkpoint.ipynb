```python
from matplotlib import pyplot as plt
x = [95,85,72,65,70]
y = ['shubham','kevin','ankit','harshid','shrey']
plt.plot(x,y)
plt.title('Result')
plt.xlabel('marks')
plt.ylabel('name')
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')

x = [100,150,200,170,125]
y = ['dell','hp','acer','mac','hcl']

plt.plot(x,y)
plt.title('Laptop price')
plt.xlabel('price')
plt.ylabel('name')
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')

x1 = [5,2,7]
y1 = [1,8,3]
x2 = [7,2,6]
y2 = [4,3,5]

plt.plot(x1,y1 ,'g',label = 'first line', linewidth = 5)
plt.plot(x2,y2 ,'r',label = 'second line', linewidth = 5)

plt.legend()
plt.grid(True ,color ='k')

plt.title('Test')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')

plt.bar([2,4,6,8],[1,2,5,3],label = 'first bar')
plt.bar([1,3,5,7],[4,6,3,1],label = 'second bar',color ='g')

plt.legend()
plt.grid(True,color = 'c')
plt.title('bar test')
plt.xlabel('X axis')
plt.ylabel('bar height')
plt.show()
```
